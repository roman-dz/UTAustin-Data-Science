```python
# Which type of book is most available?
df['Type'].value_counts()
```

```python
# Which is most commonly available Rating?
df['Rating'].value_counts()
```

```python
# Which book of Paperback type has the highest Rating 

df1 = df[df['Type'] == 'Paperback']
df1[df1['Rating']==df1['Rating'].max()]

# df1.groupby(['Book_title'])['Rating'].max().sort_values(ascending=False)
```

```python
# Which type of book has no book of price more than 150?

df2 = df[df['Price'] > 150]

df2['Type'].unique()
```

```python
# Another way to solve for -- # Which type of book has no book of price more than 150?

df.groupby(['Type'])['Price'].max().sort_values(ascending=False)
```

```python
# Which one looks similar to the distribution of Price?
sns.distplot(df["Price"],color='darkorange') 
```

```python
# Which of the following will be the countplot for Type?
sns.countplot(data = df, x='Type');
```

```python
# How will the boxplot visualization for Rating vs Type look like?
plt.figure(figsize=(15,7))           
sns.boxplot(df['Rating'],df['Type'])
plt.show()
```

```python
# Which of these will be stripplot of Rating and Type?
plt.figure(figsize=(15,7))   
sns.stripplot(data = df, x = 'Type', y='Rating', jitter=True);
```

```python
# Which of these will be KDE jointplot of Rating and price?
sns.jointplot(data=df, x="Price", y="Rating", kind="kde");
```

```python
# Which of these will be jointplot of Number_Of_Pages and price?
sns.jointplot(data=df, x="Price", y="Number_Of_Pages");
plt.show()
```

```python
sns.kdeplot(
    data=df, x="Price", y="Rating", fill=True,
)
```
