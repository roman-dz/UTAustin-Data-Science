### Practice Matplotlib

```python
x=np.arange(1,11)
y=2*x
y1=3*x
y2=4*x
x,y,y1,y2
```

```python
plt.plot(x,y,c="red",linewidth=4,linestyle=":")
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.plot(x,y1,c="g",linewidth=4,linestyle=":")
plt.plot(x,y2,c="r",linewidth=4,linestyle="-.")

plt.title('Line Plot')
plt.xlabel('Multiple plots')
plt.ylabel('Red and Green')
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,c="g",linewidth=4,linestyle=":")
plt.subplot(2,1,2)
plt.plot(x,y2,c="r",linewidth=4,linestyle="-.")
plt.show()
```

### Bar Plot

```python
student = {'Sam':67, 'Matt':23, 'Julia':93}
```

```python
Names=list(student.keys())
```

```python
Values=list(student.values())
```

```python
plt.bar(Names,Values)
plt.show()
```

```python
plt.barh(Names,Values,color="r")
plt.title('Student')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.show()
```

### Box Plot and Violin Plot

```python
one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5,4]

data = [one,two,three]
```

```python
plt.boxplot(data)
plt.show()
```

```python
plt.violinplot(data, showmedians=True)
plt.show()
```

### Pie charts

```python
fruit=['Apple','Orange','Mango','Guava']
quantity = [67,34,100,29]

plt.pie(quantity,labels=fruit)
plt.show()
```

```python
# change colors and show percentages
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['yellow','grey','blue','black'])
plt.show()
```

```python
plt.pie(quantity,labels=fruit,radius=2)
plt.pie([1],colors=['yellow'],radius=0.5)
plt.show()
```
