# Tips Dataset


`Data Features:`

**total_bill**: a numeric vector, the bill amount (dollars)


**tip**: a numeric vector, the tip amount (dollars)


**sex**: a factor with levels Female Male, gender of the payer of the bill


**smoker**: a factor with levels No Yes, whether the party included smokers


**day**: a factor with levels Friday Saturday Sunday Thursday, day of the week


**time**: a factor with levels Day Night, rough time of day


**size**: a numeric vector, number of people in a table/Party

### Q1] Import the libraries

```python
import warnings # in order to avoid warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

### Q3] Understand the distribution of 'tip' and 'total_bill' using different plots

Hint: Use distplot and boxplot

```python
sns.distplot(tips['tip'], color='green');
plt.show()
sns.boxplot(tips['tip'], color='snow');
plt.show()
```

```python
sns.distplot(tips['total_bill'], color='coral');
plt.show()
sns.boxplot(x='total_bill',data=tips, color='green');
```

### General Summary of Tip and Total Bill


- 50% of people are paying less than 17 dollars for a total bill
- 50% of people are giving less than 3-dollar tips

### Q4] Relationship between 'total_bill' and 'tip'

Implement jointplot, hex jointplot, pairplot and lmplot

```python
sns.jointplot(tips['tip'], tips['total_bill'], color='darkgreen');
```

```python
sns.jointplot(tips['tip'], tips['total_bill'], kind='hex', color='lightblue');
```

```python
sns.pairplot(tips[['total_bill','tip']]);
```

```python
sns.lmplot(x='tip',y='total_bill',data=tips);
```

### Relationship Summary

- There is a linear relationship: as total_bill amount increases so does the tip

### Q5] Make plots to check distribution of tip with different gender. 

Use striplot and boxplot

```python
sns.stripplot(tips['sex'],tips['tip']);
```

```python
sns.boxplot(tips['sex'],tips['tip']);
```

### Summary

- Median value for both genders is nearly the same
- Male tippers seem to have more outliers towards higher tip amounts

### Q6] Draw relationship between 'total_bill', 'tip' and 'sex'

use lmplot

```python
sns.lmplot(x='tip',y='total_bill',data=tips,fit_reg=False,hue='sex');
```

### Q7] Which day bring in higher bill amount?

use stripplot, swarmplot and boxplot

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False);
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips);
```

```python
sns.boxplot(x='day',y='total_bill',data=tips);
```

### Summary

- Saturday and Sunday appear as good days for restaurants due to larger number of customers and higher bill amounts.

### Q8] Find out count of total males and females visiting each day.

use countplot

```python
sns.countplot(tips['day'], hue=tips['sex']);
```

Best day of the week for single males to visit restaurants is on Thursdays.  There is the least male competition on those days.

### Q9] At what time of which day is the total bill highest with respect to tips (consider for male and female separately)

use lmplot

```python
sns.lmplot(x='tip',y='total_bill',data=tips,fit_reg=False,hue='day',row='time',col='sex');
```

Both female and male customers spend more on Saturdays and Sundays at dinner time.
